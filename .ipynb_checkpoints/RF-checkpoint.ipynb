```python
import pandas as pd
data_set = pd.read_csv("data_1.csv")
data_set.head()
```

```python
data=pd.DataFrame({
    'a':data_set.iloc[:75, 2].values,
    'b':data_set.iloc[:75, 3].values,
    'c':data_set.iloc[:75, 4].values,
    'prodi':data_set.iloc[:75, 5].values
})
data.head(75)
```

```python
import pandas as pd
data_set = pd.read_excel("data_5.xlsx")
data_set.head()
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

X=data[['a', 'b','c']]  # Features
y=data['prodi']  # Labels

#Extracting Independent and dependent Variable  
X= data_set.iloc[:75, 2:5].values  
y= data_set.iloc[:75, 5].values  

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Random Forest Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
